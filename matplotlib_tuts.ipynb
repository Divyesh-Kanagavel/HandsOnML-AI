```python
#sub plots
# plt.subplot draws on the currently active subplot -> which can be referenced by plt.gca()
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2,2,1)
plt.plot(x, x)
plt.subplot(2,2,2)
plt.plot(x,x**2)
plt.subplot(2,2,3)
plt.plot(x, x**3)
plt.subplot(2,2,4)
plt.plot(x, x**4)
plt.show()
```

```python
#more complex subplotting
x = np.linspace(-1.4, 1.4, 30)
plt.subplot2grid(shape=(3,3), loc=(0,0), rowspan = 2, colspan = 2)
plt.plot(x,x)
plt.subplot2grid(shape=(3,3), loc = (0,2))
plt.plot(x,x**2)
plt.subplot2grid(shape=(3,3), loc = (1,2), rowspan = 2)
plt.plot(x,x**3)
plt.subplot2grid(shape=(3,3), loc = (2,0), colspan = 2)
plt.plot(x, x**4)
plt.show()
```

```python
#multiple figures
#by default -> matplotlib by default assigns number 1 to figure figure(1), we can assign other numbers and work on corresponding figure
#current figure is tracked by matplotlib using the gcf()
x = np.linspace(-1.4, 1.4, 30)
plt.figure(1)
plt.figure(1)
plt.subplot(2,1,1)
plt.plot(x,x)
plt.subplot(2,1,2)
plt.plot(x,x**2)

plt.figure(2, figsize=(10,5))
plt.subplot(2,1,1)
plt.plot(x,x**3)
plt.subplot(2,1,2)
plt.plot(x,x**4)

plt.figure(1) # going back to figure 1
plt.subplot(2,1,2)
plt.plot(x, -x**3)


plt.show()
```

```python
#Drawing text

x = np.linspace(-1.5,1.5,30)
px = 0.8
py = px ** 2

plt.plot(x,x**2, 'b-', px, py, 'ro')

plt.text(0,1.5,s="Square function \n $y = x^2 $", fontsize=20, color="blue", ha = "center")

plt.text(px - 0.08, py, "Beatiful point", ha="right", weight = "heavy")
plt.text(px + 0.05, py - 0.4, s="x = %0.2f\n y = %0.2f"%(px,py), rotation=-30,color='gray')
plt.show()

```

```python
#Annotation

plt.plot(x,x**2, 'b-', px,py, 'ro')
plt.annotate("Beautiful point", xy= (px,py), xytext = (px - 1.3, py + 1.0), weight= "heavy", fontsize=14, arrowprops = {"facecolor": "lightgreen"})
plt.show()
```

```python
plt.plot(x, x**2, 'b-', px,py, 'ro')
bbox_props = dict(boxstyle="rarrow, pad=0.3", fc="lightblue", ec="b", lw=2)
plt.text(px-0.2, py, "beautiful point", ha="right", bbox = bbox_props)

bbox_props = dict(boxstyle="round4, pad=1, rounding_size=0.2", ec="black", lw=5, fc = "gray")
plt.text(0,1.5,"Square function \n $y = x^2 $", fontsize=20, color = "lightblue", ha = "center", bbox = bbox_props)
```

```python
#Non linear scale : logarithms / logits
x = np.linspace(0.01, 10, 100)
plt.figure(1)
y = x**3/np.exp(2*x)
plt.plot(x,y)
plt.yscale("linear")
plt.title("linear scale")
plt.grid(True)

# negative values invalid
plt.figure(2)
plt.plot(x,y)
plt.yscale("log")
plt.title("logarithmic scale")
plt.grid(True)

# logits - inverse function of logistic sigmoid - ln(p) / (1-1n(p))
plt.figure(3)
plt.plot(x,y)
plt.yscale("logit")
plt.title("logit scale")
plt.grid(True)

#symmetric log - handles negative values as well - positive values passed to negative values : linthresh : -val to val -> linear scale
plt.figure(4)
plt.plot(x,y - y.mean())
plt.yscale("symlog", linthresh = 0.05)
plt.title("symlog scale")
plt.grid(True)

plt.show()

```

```python
#ticks and tickers
x = np.linspace(-2,2,100)
plt.figure(1, figsize=(15,10))
plt.subplot(131)
plt.plot(x,x**3)
plt.grid(True)
plt.title("Default ticks")

ax = plt.subplot(132)
ax.plot(x,x**3)
ax.xaxis.set_ticks(np.arange(-2,2))

ax.grid(True)
plt.title("Custom ticks 1")

ax = plt.subplot(133)
ax.plot(x,x**3)
ax.xaxis.set_ticks([-2,0,1,2])
ax.yaxis.set_ticks(np.arange(-5,5,1))
ax.yaxis.set_ticklabels(["min", -4, -3, -2, -1, 0, 1, 2, 3, "max"])
ax.grid(True)
plt.title("custom ticks 2")

plt.show()

```

```python
#polar projection - an alternative to cartesian coordinate system

radius = 1
theta = np.linspace(0,2*np.pi, 100)
plt.subplot(111, projection="polar")
plt.plot(theta, np.sin(5*theta), "g-")
plt.plot(theta, 0.5*np.cos(20*theta), "b-")
plt.show()
```

```python
#3d projection
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X,Y = np.meshgrid(x,y)
Z = np.sqrt(X**2 + Y**2)

R = np.sin(Z)

plt.figure(1, figsize=(12,4))
subplot3d = plt.subplot(111, projection="3d")
subplot3d.plot_surface(X,Y,R, cstride=1, rstride=1, cmap=matplotlib.cm.coolwarm, linewidth = 0.1)
plt.show()
```

```python
# contour of the same data
plt.contourf(X, Y, Z, cmap = matplotlib.cm.coolwarm)
plt.show()
```

```python
#Lines plots
def plot_lines(axis, slope, intercept, **kwargs):
  xmin, xmax = axis.get_xlim()
  axis.plot([xmin, xmax], [xmin*slope + intercept, xmax * slope + intercept], **kwargs)

x = np.random.randn(1000)
y = 0.5 * np.random.randn(1000) + 0.25 * x + 0.2

plt.axis([-3.5,3.5,-3.5,3.5])
plt.scatter(x,y,alpha = 0.2)
plt.plot(0,0,'ro')

plt.hlines(0, -3.5, 0, color="red")
plt.vlines(0, -3.5,0, color="red")
plot_lines(axis=plt.gca(), slope=0.25, intercept=0.2, color="magenta")
plt.grid(True)
plt.show()





```
