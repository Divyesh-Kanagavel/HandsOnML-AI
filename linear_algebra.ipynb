Dot product - an important idea in linear algebra. Also called as scalar product or inner product

$\textbf{u} \cdot \textbf{v} = ||\textbf{u}|| \times ||\textbf{v}|| cos(\theta)$\

Given two vectors u and v, it is computed as \
$\textbf{u} \cdot \textbf{v} = \Sigma (u_i \times v_i)\$

```python
#implementation of dot product in python
def dot_product(v1,v2):
  return sum([v1i*v2i for v1i, v2i in zip(v1,v2)])

dot_product(u,v)
```

```python
#efficient way is to use numpy or pytorch
np.dot(u,v)
```

```python
#dot method of numpy arrays
u.dot(v)
```

```python
# * operator will perform elementwise product called hadamard product and not dot product of vectors
u * v
```

Properties of dot product:
commutative : u . v = v . u\
Dot product is defined between two vector and not between scalar and vector\
Hence, dot product is not associative, since dot of two vectors produces a scalar and cannot be dotted with another vector.\
But dot product is associative w.r.t a scalar, that is, \
 $\lambda \times (u \cdot v)$ = $(\lambda u) \cdot v$ = $(u) \cdot \lambda v$\
 Dot product is distributive,\
 $u \cdot (v+w)$ = $ (u \cdot v) + (u \cdot w)$

Another use case of dot product is to calculate the angle between two vectors given by : \
$\theta = \arccos{\frac{u \cdot v}{||u|| \times ||v||}}$
If $u \cdot v $ = 0, $\theta = \frac{\pi}{2}$

```python
#python code to calculate angle between two vectors
def angle_vectors(v1,v2):
  cos_theta = np.dot(v1,v2)/linalg.norm(v1)/linalg.norm(v2)
  return np.arccos(cos_theta.clip(-1,1))
```

```python
angle = angle_vectors(u,v)
```

```python
print(angle)
```

```python
def plot_projection(U, P):
    U_P = U @ P

    axis_end = 100 * U
    plot_vector2d(axis_end[0], color="black")

    plt.gca().add_artist(Polygon(P.T, alpha=0.2))
    for vector, proj_coordinate in zip(P.T, U_P.T):
        proj_point = proj_coordinate * U
        plt.plot(proj_point[0][0], proj_point[0][1], "ro", zorder=10)
        plt.plot([vector[0], proj_point[0][0]], [vector[1], proj_point[0][1]],
                 "r--", zorder=10)

    plt.axis([0, 5, 0, 4])
    plt.gca().set_aspect("equal")
    plt.grid()
    plt.show()
```

```python
#consider this matrix which contains elements which are related to a parameter theta = 30 degrees
angle30 = 30 * np.pi/180
rot_matrix = np.array([
                       [np.cos(angle30), np.sin(angle30)],
                       [-np.sin(angle30), np.cos(angle30)]
])

rot_matrix
```

```python
rot_matrix @ P
```

The new matrix obtained by the mat mul above is basically rotating the matrix P by an angle of -30 around origin. the horizontal and vertical axes are rotated by 30 degrees about the origin throught his matrix multiplication (projection of each of the axis on the axis specified by rot_matrix)

```python
def plot_transformation(P_before, P_after, text_before, text_after, axis = [0, 5, 0, 4], arrows=False):
    if arrows:
        for vector_before, vector_after in zip(P_before.T, P_after.T):
            plot_vector2d(vector_before, color="blue", linestyle="--")
            plot_vector2d(vector_after, color="red", linestyle="-")
    plt.gca().add_artist(Polygon(P_before.T, alpha=0.2))
    plt.gca().add_artist(Polygon(P_after.T, alpha=0.3, color="r"))
    plt.plot(P_before[0], P_before[1], "b--", alpha=0.5)
    plt.plot(P_after[0], P_after[1], "r--", alpha=0.5)
    plt.text(P_before[0].mean(), P_before[1].mean(), text_before, fontsize=18, color="blue")
    plt.text(P_after[0].mean(), P_after[1].mean(), text_after, fontsize=18, color="red")
    plt.axis(axis)
    plt.gca().set_aspect("equal")
    plt.grid()
```

```python
P_rotated = rot_matrix @ P
```

```python
plot_transformation(P, P_rotated, "$P$", "$VP$", [-2, 6, -2, 4], arrows=True)
plt.show()
```

Properties:\

$(\lambda \times M)^{-1} = \frac{1}{\lambda} \times M^{-1}$, where $\lambda$ is a scalar quantity.

A matrix that is its own inverse is called an involution. The simplest examples are reflection matrices, or a rotation by 180°, but there are also more complex involutions, for example imagine a transformation that squeezes horizontally, then reflects over the vertical axis and finally rotates by 90° clockwise

Square matrix whose inverse is its own transpose is called orthogonal matrix.\
$H^{-1} = H^T$\

Singular value decomposition (SVD) : these matrix manipulations and decompositions are extensively used to solve linear algebra equations efficienlty converting a complex matrix to bunch of smaller more expressible matrices with interesting properties thereby enabling faster solutions to equations

Any $m \times n$ matrix can be decomposed into three matrices:\
1. an orthogonal rotation $m \times m$ matrix U\
2. a diagonal matrix S_diag of shape $m \times n $ - a scaling and projecting matrix\
3. another orthogonal matrix of shape $n \times n$\
M = U S $V^T$
